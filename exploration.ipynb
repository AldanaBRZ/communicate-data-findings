## Bivariate Exploration

> Now I will explore the relationship between school grade and wealth. I want to see if they are related. First I will do a scatterplot with some jitter to try and see the distribution of the data.

```python
sb.regplot(data = df_clean, x = 'grade', y = 'wealth', truncate = False, x_jitter = 0.3,scatter_kws={'alpha':1/20})
plt.xlabel('Grade')
plt.ylabel('Wealth')
plt.title('Regplot of wealth against grade');
```

> The distribution is not very clear because there are a lot of data points, but we can still get some insights. The most common school grade is the average grade, as expected. There is also a clear trend upwards, which means there is a positive correlation between wealth and grade (as one goes up, the other one goes up as well). This implies that welthier children go to schools where the grade is higher.

> Maybe a violin plot will show us a clearer representation of how the data is distributed. 

```python
sb.violinplot(data = df_clean, x = 'grade', y = 'wealth', color = base_color, inner = None)
plt.xlabel('Grade')
plt.ylabel('Wealth')
plt.title('Violinplot of wealth against grade');
```

> We can see that the bulk of the data in each grade goes up in wealth as the grade goes up. That is, except on the highest grade. There the bulk on the data is lower than expected. That could be due to there being less data points there (less schools are so demanding, so that data is not as representative as it could be).

> I will now make a boxplot of the same columns to see if/ where there are outliers and where are the medians of the grades.

```python
plt.figure(figsize = (12,7)) 
sb.boxplot(data = df_clean, x = 'grade', y = 'wealth', color = base_color)
plt.xlabel('Grade', fontsize = 13)
plt.ylabel('Wealth', fontsize = 13)
plt.xticks(fontsize = 10)
plt.yticks(fontsize = 10)
plt.title('Boxplot of wealth against grade', fontsize = 15);
```

> We can see that there are a lot of outliers, and we can arrive to the same conclusions we arrived with the violin plot. There is a clear upwards trend, with the medians as well, and the highest grade goes down abruptly in what regards to wealth.

> I will now make some comparisons with the exams grades and wealth.

```python
plt.figure(figsize = (20, 5))

plt.subplot(1, 3, 1)
sb.regplot(data = df_clean, x = 'wealth', y = 'grades_math', scatter_kws = {'alpha' : 1/20})
plt.ylabel('Math grades')
plt.xlabel('Wealth')
plt.title('Math grades against wealth');


plt.subplot(1, 3, 2)
sb.regplot(data = df_clean, x = 'wealth', y = 'grades_read', scatter_kws = {'alpha' : 1/20})
plt.ylabel('Reading grades')
plt.xlabel('Wealth')
plt.title('Reading grades against wealth');

plt.subplot(1, 3, 3)
sb.regplot(data = df_clean, x = 'wealth', y = 'grades_science', scatter_kws = {'alpha' : 1/20})
plt.ylabel('Science grades')
plt.xlabel('Wealth')
plt.title('Science grades against wealth');
```

> The scatterplots don't tell us much because there are a lot of points. Still, we can see an upwards trend. Those that are wealthier tend to have better grades. I will do heat maps to see if we can make sense of what the distributions are. 

```python
fig, ax = plt.subplots(1, 3, figsize = (20, 5))
bins_x = np.arange(-7, 3+0.6, 0.6)
bins_y = np.arange(0, 1000+40, 40)

plt.subplot(1, 3, 1)
plt.hist2d(data = df_clean[['wealth', 'grades_math']].dropna() , x = 'wealth', 
           y = 'grades_math', cmap = 'viridis_r', cmin = 10, 
           bins = [bins_x, bins_y], vmin = 0, vmax = 18000)
plt.ylabel('Math grades', fontsize = 15)
plt.xlabel('Wealth', fontsize = 15)
plt.title('Math grades against wealth', fontsize = 15)
plt.xticks(fontsize = 15)
plt.yticks(fontsize = 15)

plt.subplot(1, 3, 2)
plt.hist2d(data = df_clean[['wealth', 'grades_read']].dropna() , x = 'wealth', 
           y = 'grades_read', cmap = 'viridis_r', cmin = 10, 
           bins = [bins_x, bins_y], vmin = 0, vmax = 18000)
plt.ylabel('Reading grades', fontsize = 15)
plt.xlabel('Wealth', fontsize = 15)
plt.title('Reading grades against wealth', fontsize = 15)
plt.xticks(fontsize = 15)
plt.yticks(fontsize = 15)

plt.subplot(1, 3, 3)
plt.hist2d(data = df_clean[['wealth', 'grades_science']].dropna() , 
           x = 'wealth', y = 'grades_science', cmap = 'viridis_r', 
           cmin = 10, bins = [bins_x, bins_y], vmin = 0, vmax = 18000)
plt.ylabel('Science grades', fontsize = 15)
plt.xlabel('Wealth', fontsize = 15)
plt.title('Science grades against wealth', fontsize = 15)
plt.xticks(fontsize = 15)
plt.yticks(fontsize = 15)

im = plt.gca().get_children()[0]
cax = fig.add_axes([1, 0.1, 0.03, 0.85])
cbar = fig.colorbar(im, cax = cax)
cbar.ax.tick_params(labelsize=15);
```

> We can see that the three exams grades have very similar distributions. Most of the data is in the average wealth (0) and between 400 and 600 in grades.
